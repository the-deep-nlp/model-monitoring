```python
drift_profile = Profile(sections=[CatTargetDriftProfileSection()])
```

```python
df.project_id.unique()
```

```python
results = []
```

```python
for ref_project_id in df.project_id.unique():
    for prod_project_id in df_new_data.project_id.unique():
        if ref_project_id == prod_project_id:
            df_reference = use_mask_project_id_based(df, project_id=ref_project_id)
            df_reference = df_reference.drop(["project_id", "entry_id", "excerpt"], axis=1)
            df_prod = use_mask_project_id_based(df_new_data, project_id=prod_project_id)
            df_prod = df_prod.drop(["project_id", "entry_id", "excerpt"], axis=1)

            drift_profile.calculate(df_reference, df_prod, column_mapping=None)
            drift_profile_json = json.loads(drift_profile.json())
            target_drift = drift_profile_json["cat_target_drift"]["data"]["metrics"]["target_drift"]

            results.append({
                "reference_project_id": ref_project_id,
                "prod_project_id": prod_project_id,
                "target_drift": target_drift
            })
```

```python
results_df = pd.DataFrame.from_records(results)
```
